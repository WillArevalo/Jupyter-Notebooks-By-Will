# Variables

Para que un modelo pueda ser entrenable es necesario que las salidas puedan ser modificables pero con las mismas entradas
Esto permiten las variables

```
A = tf.Variable([2.0],dtype=tf.float32)
B = tf.Variable([-1.0], dtype=tf.float32)
x = tf.placeholder(tf.float32)

#Operacion y salida
salida = A * x + B

#Inicializar variables
init = tf.global_variables_initializer()

#Creamos la sesion
sess = tf.Session()
#Se corre init para inicializar las variables
sess.run(init)

#Imprimimos
print("A y B :",sess.run([A,B]))
print("Ejemplo A=2, x=3, B=-1; A*x+B = 5")
print("Resultado de A * x + B =",sess.run(salida,{x:[1,2,3,4]}))
```

## Modificando variables

```
#Se pueden modificar facilmente con el metodo assign y se envian como parametros el objeto a modificar y como se modificara

Ax = tf.assign(A,A+1) #Ax = 2+1 = 3  
Bx = tf.assign(B,B-2) #Bx = -1-2 = -3

```

Para que estos valores realmente se modifiquen se deben correr dentro de la sesion

```
#Imprimimos
print("A y B :",sess.run([Ax,Bx]))
print("Ejemplo A=3, x=3, B=-3; A*x+B = 6")
print("Resultado de A * x + B =",sess.run(salida,{x:[1,2,3,4]}))
```

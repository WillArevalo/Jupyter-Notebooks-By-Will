# Operaciones entre dos tensores

```
node1 = tf.constant(7.0, dtype=tf.float32)
node2 = tf.constant(5.0)

sess = tf.Session()
```

## Operacion Suma

```
node3 = tf.add(node1,node2)
print("node3 =",sess.run(node3))
```

## Marcadores de posicion (placeholder)
Estos son parametros que aceptan entradas externas, lo que indica que se le puede asignar un valor mas adelante

```
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
sumador = x + y

print("",x,"\n",y,"\n",sumador)
```

```
print("sumador #1 :",sess.run(sumador,feed_dict={x:3,y:2}))
print("sumador #1 :",sess.run(sumador,{x:[2,3],y:[4,1]}))
```

```
#a√±adiendo operaciones
sum_mul = sumador * 5
print("sum_mul5 #1 :",sess.run(sum_mul,{x:3,y:4}))
```
